```python
import pandas as pd
import matplotlib as plt
import re
```

```python
dsjobs = pd.read_csv("sd_jobs.csv")
dsjobs_description = dsjobs["job_description"]
dsjobs_description.count()
```

```python
years = []

for desc in dsjobs_description:
    split_text = desc.split()
    if "year" in split_text:
        index = split_text.index("year")
        parsed_text = (' '.join(split_text[max(0,index-2):min(index+2,len(split_text))]))
        text_nums = re.findall('\d+', parsed_text)
        
#  Conditional Omits
        if '$' not in parsed_text:
            print(parsed_text)
        for x in text_nums:
            if int(x) < 10:
                print(x)
                years.append(int(x))
        print("\n")
        
            
```
